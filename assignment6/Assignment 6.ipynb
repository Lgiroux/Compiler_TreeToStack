```python
from cuppa1_lex import lexer
from cuppa1_frontend_gram import parser
from cuppa1_state import state
from cuppa1_stack_codegen import walk as codegen
from cuppa1_cc_output import output
from pprint import pprint
from stackmachine_interp import interp
```

code refers to the code that is generated by the compiler
it is originally code used by a tree-walking interpreter but the compiler
modifies it to be interpreted by a stack interpreter

```python
def compiler(input_stream):
    parser.parse(input_stream, lexer=lexer)
    tuples = codegen(state.AST) + [('stop',)]
    code = output(tuples)
    return code
```

```python
input_stream = \
'''
x = 1;
put x;
'''
```

```python
parser.parse(input_stream, lexer=lexer)
```

```python
code = compiler(input_stream)
```

```python
print(code)
```

```python
interp(code)
```

```python
input_stream = \
'''
get x;
x = x + 1;
put x;
'''
parser.parse(input_stream, lexer=lexer)
code = compiler(input_stream)
```

```python
print(code)
interp(code)
```

```python
input_stream = \
'''
get x;
if (x)
    put 1;
else
    put 0;
'''
parser.parse(input_stream, lexer=lexer)
code = compiler(input_stream)
```

```python
print(code)
interp(code)
```
